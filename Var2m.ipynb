```
#Momentum GD 


#class function for function and 1st derivative definition 
class F(object):
  def __init__(self):
    pass
  
  def func(self, x1, x2): 
    y = 2*x1**2-1.05*x1**4+x1**6/6+x1*x2+x2**2
    return y
  
  def deriv(self, x1, x2): #for gradient calculation 
    dx1 = 4*x1-4.2*x1**3+x1**5+x2 
    dx2 = x1+2*x2
    return dx1, dx2
  
#define momentum GD class + function 
class grad_desc(object):
  def __init__(self, function, alpha = 0.00215, n_iter = 1000, tol = 1e-5):
    self.alpha = alpha
    self.n_iter = n_iter
    self.tol = tol
    self.function = function
  
  def grad_desc(self, x1, x2):
    
    dx1, dx2 = self.function.deriv(x1, x2)
   
   #momentum algorithm
    new_x1 = x1 + (0.5 * (self.alpha * dx1) ** 2) - (self.alpha * dx1)
    new_x2 = x2 + (0.5 * (self.alpha * dx2) ** 2) - (self.alpha * dx2)
    
    return new_x1, new_x2, dx1, dx2
  
  def optimise(self, start_x1 = 1, start_x2 = 1):
    all_x1 = [start_x1]
    all_x2 = [start_x2]
    all_y = [self.function.func(start_x1, start_x2)]
    
    
    for i in range(self.n_iter):
      x1,x2,dx1,dx2 = self.grad_desc(all_x1[i], all_x2[i])
      
      all_x1.append(x1)
      all_x2.append(x2)
      y = self.function.func(x1, x2)
      all_y.append(y)
      
      print("i", i, "x1", x1,"x2", x2,"y", y)
      
    
    return all_x1, all_x2, all_y
#call function 
function1 = F()
gd = grad_desc(function = function1)
#output GD path
all_x1, all_x2, all_y = gd.optimise()

```
