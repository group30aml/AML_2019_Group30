```
#function 
import numpy as np

class f_y(object):
  def __init__(self):
    pass
  
  def evaluate_function(self, x1, x2): 
    y = 2*x1**2-1.05*x1**4+x1**6/6+x1*x2+x2**2
    return y
  
  def evaluate_derivative(self, x1, x2): #for gradient calculation 
    dx1 = 4*x1-4.2*x1**3+x1**5+x2 
    dx2 = x1+2*x2
    return dx1, dx2
  
  
    
#class for stochastic GD calculations
class grad_desc_alg(object):
  def __init__(self, function, learn_rate = 0.0001, max_epoch = 10000, thres = 1e-5):
    self.learn_rate = learn_rate
    self.max_epoch = max_epoch
    self.thres = thres
    self.function = function
    
   
    def x1_shuffled(self,x1):
      x1_shuffled = np.random.shuffle(x1)
      return x1_shuffled
    def x2_shuffled(self,x2):
      x2_shuffled = np.random.shuffle(x2)
      return x2_shuffled
 
  
  
  def grad_desc(self, x1_shuffled, x2_shuffled):
    
    dx1, dx2 = self.function.evaluate_derivative(x1_shuffled, x2_shuffled)
    
#stochastic GD
      
    new_x1 = x1_shuffled - self.learn_rate * dx1
    new_x2 = x2_shuffled - self.learn_rate * dx2
    
    return new_x1, new_x2, dx1, dx2
  
  def optimise(self, start_x1 = -9, start_x2 = 1):
    all_x1 = [start_x1]
    all_x2 = [start_x2]
    all_y = [self.function.evaluate_function(start_x1, start_x2)]
    
    
    for epoch in range(self.max_epoch):
      x1_shuffled,x2_shuffled,dx1,dx2 = self.grad_desc(all_x1[epoch], all_x2[epoch])
      
      all_x1.append(x1_shuffled)
      all_x2.append(x2_shuffled)
      y = self.function.evaluate_function(x1_shuffled, x2_shuffled)
      all_y.append(y)
      
      print("epoch", epoch, "x1", x1_shuffled,"x2", x2_shuffled,"y", y)
      
      if (np.abs(dx1) + np.abs(dx2)  < 2*self.thres):
        break
    
    return all_x1, all_x2, all_y

function1 = f_y()
gd = grad_desc_alg(function = function1)

#output path for stochastic GD
all_x1, all_x2, all_y = gd.optimise()


```
